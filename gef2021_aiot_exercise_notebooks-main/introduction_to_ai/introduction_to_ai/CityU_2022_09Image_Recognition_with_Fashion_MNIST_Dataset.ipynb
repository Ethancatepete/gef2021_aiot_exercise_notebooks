# **Model training**

### Softmax function



At the output layer, the softmax function is used to normalize the output scores into probabilities:


$$p_i = softmax(x_i) = \frac{\exp({y_i})}{ \sum_{j=1}^n \exp({y_j})}$$

After applying the softmax function, the sum of the output probabilities will be 1:
 $$\sum_{i=1}^n p_i = 1$$




```python
#Type your code here

# Reshaping the data
INPUT_SHAPE = [28*28]   #input shape for one image

x_train = x_train.reshape([-1]+INPUT_SHAPE)
x_test = x_test.reshape([-1]+INPUT_SHAPE)

print(f'Train data shape: {x_train.shape}, Test data shape: {x_test.shape}')
```

```python
from tensorflow.keras.layers import *


tf.keras.backend.clear_session() #to clear previous runs 

#Type your code here

#hyper parameters - to be tunned
#Define and compile your model - Defining the Nurual Network - using Functional API (which is supported on FPGA and allows for more complex architecture compared to the sequencial functions)
#=============================================
inputs = tf.keras.Input(shape=INPUT_SHAPE)
x = Dense(units=100, activation='relu')(inputs) #hidden layer, inputs is the input for this model, input (wihtout the s) is a reserved term
#the num of params in dense is 78500 because it is 784*100 + 100 (error term? something like that)
#x = Dense(units=100, activation='relu')(x) to add another layer, you can override previous vars, unless you are using res-net 
outputs = Dense(units=num_classes, activation='softmax')(x) # x is the input for this layer


model = tf.keras.Model(inputs=inputs, outputs=outputs, #there can me multi-in multi-out models, just make a list for inputs and outputs, eg: inputs = [x,y]
                       name="dnn_mnist_model"
                       )

#optimizer = tf.keras.optimizers.SGD(learning_rate=0.01,name='SGD') #https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/SGD

optimizer = tf.keras.optimizers.Adam(
    learning_rate=0.001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=False,
    name='Adam'
)

model.compile(
    loss = 'catagorical_crossentropy', #the loss function is defined by keras already 
    optimizer=optimizer,
    metrics=["accuracy"]
)


#=============================================
model.summary()
#output shape is none,784 etc because we didn't define the no. of params yet
```
